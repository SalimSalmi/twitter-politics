```python
import pandas as pd
weeks = pd.read_csv("/Users/user/Delft/NN/proj/weeklyTimeline.csv")
```

```python
pytrend.build_payload(kw_list=['bitcoin'],timeframe='2012-04-08 2012-04-14')
interest_over_time_df = pytrend.interest_over_time()
week1 = interest_over_time_df; week1
```

```python
pytrend.build_payload(kw_list=['bitcoin'],timeframe='2012-04-15 2012-04-21')
interest_over_time_df = pytrend.interest_over_time()
week2 = interest_over_time_df; week2
```

```python
df1 = pd.concat([week1, week2]); #twoweeks
index = df1.index.values
df2 = pd.DataFrame({"Weekly": [int(weeks[0:1]["bitcoin: (Worldwide)"])]+[""]*6+[int(weeks[1:2]["bitcoin: (Worldwide)"])]+[""]*6}, index=index)
combo = pd.concat([df1, df2], axis=1)
# df1.index.values == df2.index.values
combo
```

```python
adjFactor = []
for x in xrange(len(combo)):
    if combo.iloc[x]["Weekly"] != '':
        current = float(combo.iloc[x]["Weekly"])/combo["bitcoin"][x]
        
    adjFactor.append(current)
adjFactorCol = pd.DataFrame({"AdjFactor":adjFactor}, index=index)
adjFactorCol
step3 = pd.concat([combo, adjFactorCol], axis=1)
step3
# combo.iloc[0]["Weekly"]
```

```python
adjVal = []
for x in xrange(len(step3)):
    adjVal.append(step3["bitcoin"][x]*step3["AdjFactor"][x])

adjValCol = pd.DataFrame({"AdjValue":adjVal}, index=index)
# adjValCo
step4 = pd.concat([step3, adjValCol], axis=1)
step4
# combo.iloc[0]["Weekly"]
```
